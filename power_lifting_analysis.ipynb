```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../datasets/powerlifting_dataset.csv")
df
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
print("Basic statistics for Amount Lifted in Kg : ")
df['Amount Lifted (kg)'].describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(30, 6))
plt.scatter(df['Age'], df['Amount Lifted (kg)'], color='blue', alpha=0.6)

# Adding title and labels
plt.title('Scatter Plot of Amount Lifted vs. Age')
plt.xlabel('Age')
plt.ylabel('Amount Lifted (kg)')

# Display the plot
plt.show()
```

```python
df['Age Group'] = pd.cut(df['Age'], bins=[0, 20, 30, 40, 50, 60, 70], labels=['<20', '20-29', '30-39', '40-49', '50-59', '60+'])

# Plot boxplot
plt.figure(figsize=(12, 6))
sns.boxplot(x='Age Group', y='Amount Lifted (kg)', data=df)

# Adding title and labels
plt.title('Box Plot of Amount Lifted by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Amount Lifted (kg)')

plt.show()
```

```python
top_10_strong = df.sort_values(by='Amount Lifted (kg)', ascending=False).head(10)
top_10_strong
```

```python
df['Lift Type'].unique()
```

```python
bench_press_top5 = df[df['Lift Type'] == 'Bench Press'].sort_values(by='Amount Lifted (kg)', ascending=False).head(5)[['Lifter Name', 'Age', 'Weight Class', 'Amount Lifted (kg)']]

# Filter and sort for Squat
squat_top5 = df[df['Lift Type'] == 'Squat'].sort_values(by='Amount Lifted (kg)', ascending=False).head(5)[['Lifter Name', 'Age', 'Weight Class', 'Amount Lifted (kg)']]

# Filter and sort for Deadlift
deadlift_top5 = df[df['Lift Type'] == 'Deadlift'].sort_values(by='Amount Lifted (kg)', ascending=False).head(5)[['Lifter Name', 'Age', 'Weight Class', 'Amount Lifted (kg)']]

# Display the tables
print("Top 5 Bench Press Lifters:")
print(bench_press_top5)
print()
print("Top 5 squats Lifters:")
print(squat_top5)
print()
print("Top 5 deadlift Lifters:")
print(deadlift_top5)
```

```python
# Plotting the scatter plot
plt.figure(figsize=(12, 8))

# Scatter plot for Bench Press
plt.scatter(bench_press_top5['Weight Class'], bench_press_top5['Age'], color='red', s=100, label='Bench Press')

# Scatter plot for Squat
plt.scatter(squat_top5['Weight Class'], squat_top5['Age'], color='blue', s=100, label='Squat')

# Scatter plot for Deadlift
plt.scatter(deadlift_top5['Weight Class'], deadlift_top5['Age'], color='green', s=100, label='Deadlift')

# Adding labels and title
plt.xlabel('Weight Class')
plt.ylabel('Age')
plt.title('Top 5 Lifters by Lift Type to lift 398-399kgs')
plt.legend()

# Display the plot
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.histplot(df['Age'], bins=20, kde=True, color='salmon')
plt.title('Age Distribution of Lifters')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python

plt.figure(figsize=(12, 6))
x=df['Lift Type']
y=df['Amount Lifted (kg)']
sns.barplot(x,y, estimator=np.mean, palette='viridis')

# Add a title, labels, and grid to the chart
plt.title('Average Amount Lifted by Lift Type')
plt.xlabel('Lift Type')
plt.ylabel('Average Amount Lifted (kg)')
plt.grid(True)

# Show the chart
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.countplot(x='Weight Class', data=df)
plt.title('Weight Class Distribution')
plt.show()
```

```python
plt.figure(figsize=(15, 5))
sns.boxplot(x='Weight Class', y='Amount Lifted (kg)', hue='Lift Type', data=df)
plt.title('Box plot of Amount Lifted (kg) by Weight Class and Lift Type')
plt.show()
```
